```python
def plot_vardensity_awake(df_):
    df = df_.copy()
    awake = df.loc[df.Y==0]
    sleep = df.loc[df.Y !=0]
    for k in df.iloc[:, :-1].columns:
        fig, ax = plt.subplots(figsize=(20,9))
        sns.kdeplot(awake[k], shade=True, color="r")
        sns.kdeplot(sleep[k], shade=True, color="b")
        plt.figure()
```

```python
plot_vardensity_awake(df.iloc[:, -8:])
```

```python
plot_vardensity_awake(df)
```
